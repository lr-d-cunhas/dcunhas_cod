```python
import pandas as pd
import matplotlib.pyplot as plt
```

Dados extraidos de outro código feito no vscode

```python
arquivo = pd.read_csv('/content/drive/MyDrive/random project/tabela1.csv', sep=',', encoding='ISO-8859-1')
```

```python
arquivo = arquivo[['indice', 'ac_treino', 'ac_teste', 'perc_posi', 'perc_posi_teste']].set_index('indice')
```

```python
arquivo
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.suptitle('RESULTADO MODELO EM DIFERENTES INTERVALOS DE 10 ANOS', fontsize=25)
plt.title('PERCENTUAL DIAS POSITOVOS', loc='left', fontsize=20)
arquivo['perc_posi'].plot(label='Percentual dias positivos toda base', color='gray')
plt.axhline(arquivo['perc_posi'].median(), color='black', linewidth=3, linestyle='--')
plt.legend(loc='upper left', fontsize=(12))
plt.yticks(fontsize=12)
plt.xticks(rotation=30, fontsize=12)
plt.xticks(fontsize=0,  color='w')
plt.xlabel(None)

plt.subplot(2, 2, 2)
arquivo['perc_posi_teste'].plot(label='Percentual dias positivos teste', color='gray')
plt.axhline(arquivo['perc_posi_teste'].median(), color='black', linewidth=3, linestyle='--')
plt.legend(loc='upper left', fontsize=(12))
plt.yticks(fontsize=12)
plt.xticks(fontsize=0,  color='w')
plt.xlabel(None)

plt.subplot(2, 2, 3)
plt.title('ACURÁCIA TREINO E TESTE', loc='left', fontsize=20)
arquivo['ac_teste'].plot(label='Acurácia teste', color='gray')
plt.axhline(arquivo['ac_teste'].median(), color='black', linewidth=3, linestyle='--')
plt.legend(loc='upper left', fontsize=(12))
plt.yticks(fontsize=12)
plt.xticks(rotation=30, fontsize=10)
plt.xlabel(None)

plt.subplot(2, 2, 4)
arquivo['ac_treino'].plot(label='Acurácia treino', color='gray')
plt.axhline(arquivo['ac_treino'].median(), color='black', linewidth=3, linestyle='--')
plt.legend(loc='upper left', fontsize=(12))
plt.yticks(fontsize=12)
plt.xticks(rotation=30, fontsize=10)
plt.xlabel(None)

plt.show()
```
